```python
def fact(n):
    if(n==1):
        return 1
    else:
        return n*fact(n-1)

fact(3)
```

```python
def fact(n):
    if(n==1):
        print(n)
        return 1
    else:
        print(n,"*",end=" ")
        return n*fact(n-1)

fact(5)
```

```python
def fib(n):
    if n==0 or n==1:
        return n
    return fib(n-1)+fib(n-2)

fib(3)
```
