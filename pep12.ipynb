```python
n = 5
pattern = lambda x: print(*range(1, x + 1))
list(map(pattern, range(1, n + 1)))
```

```python
fun2=lambda x,y : x if x>y else y
fun2(5,2)
```

```python
fun3=lambda x,y,z : x+y+z
fun3(1,2,3)
```

```python
pattern= lambda n: list(map(lambda i: print("*" * i), range(1,n+1)))
pattern(5)
```

```python
pattern= lambda n: list(map(lambda i: print("*" * i), range(n,0,-1)))
pattern(5)
```

```python
n = 4

list(map(lambda i: print(" " * (n - i) + " * " * i), range(1, n + 1)))
```
